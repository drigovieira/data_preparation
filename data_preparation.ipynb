```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('DataPrepFinal.csv')
```

```python
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0', 'name', 'category', 'goal', 'pledged', 'usd pledged'])
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['usd_goal_real'] = df['usd_goal_real'].apply(lambda x: x.strip('USD '))
```

```python
df.head()
```

```python
df['launched'].str.replace(' \d\d:\d\d', '', regex=True)
```

```python
df.head()
```

```python
df['launched'] = df['launched'].str.replace(' \d\d:\d\d', '', regex=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['usd_goal_real'] = df['usd_goal_real'].astype('int64')
```

```python
df.dtypes
```

```python
df['deadline'] = pd.to_datetime(df['deadline'], format='%d/%m/%y')
df['launched'] = pd.to_datetime(df['launched'], format='%d/%m/%y')
```

```python
df.dtypes
```

```python
df.head()
```

```python
df['time_range'] = (df['deadline'] - df['launched']).dt.days
```

```python
df.head()
```

```python
df_i = pd.read_csv('invested.csv')
```

```python
df_i.head()
```

```python
df_i.sort_values(by=['ID'])
```

```python
df_i.head()
```

```python
df_i = pd.get_dummies(df_i, columns=['backedLocation'])
```

```python
df_i.sort_values(by=['ID'])
```

```python
df_i = df_i.groupby(by=['ID']).agg({'age':'mean', 'backedLocation_BR':'sum', 'backedLocation_GBK':'sum', 'backedLocation_US':'sum'})
```

```python
df = df.merge(df_i, how='left', on=['ID'])
```
